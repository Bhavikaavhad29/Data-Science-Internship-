```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.arange(0, 11)
y = 3*x + 2
plt.plot(x, y); plt.show()


u = np.random.rand(50)
v = np.random.rand(50)
plt.scatter(u, v, s=50, color='r'); plt.show()


n = ['A', 'B', 'C', 'D', 'E']
w = [5, 8, 6, 9, 4]
plt.bar(n, w)
for i, j in enumerate(w):
    plt.text(i, j, str(j), ha='center')
plt.show()


t = np.linspace(0, 5, 100)
plt.plot(t, np.sin(t))
plt.plot(t, np.cos(t))
plt.plot(t, np.exp(t))
plt.show()


h = np.random.randn(1000)
plt.hist(h, bins=30, color='g'); plt.show()

# 6 Subplots
fig, ax = plt.subplots(2, 2)
ax[0,0].plot(x, y)
ax[0,1].bar(n, w)
ax[1,0].scatter(u, v)
ax[1,1].hist(h, bins=20)
plt.show()


m = np.arange(1, 13)
p1 = np.random.randint(10, 20, 12)
p2 = np.random.randint(15, 25, 12)
p3 = np.random.randint(12, 18, 12)
plt.stackplot(m, p1, p2, p3); plt.show()

d = pd.DataFrame({
    'M': np.random.randint(50, 100, 10),
    'S': np.random.randint(50, 100, 10),
    'E': np.random.randint(50, 100, 10)
})
d.boxplot(); plt.show()


lbl = ['X', 'Y', 'Z', 'W']
val = [40, 25, 20, 15]
exp = [0.1 if v == max(val) else 0 for v in val]
plt.pie(val, labels=lbl, explode=exp, autopct='%1.1f%%'); plt.show()

 
px = np.arange(10)
py = np.random.randint(10, 50, 10)
plt.plot(px, py, label='Data')
plt.gca().invert_xaxis()
plt.legend(); plt.show()
```

```python
years=['2001','2003','2005','2007','2009']
sachin=[1250,980,1130,1020,870]
kohli=[0,0,430,880,1050]
sehwag=[950, 1230, 890, 970, 920]
x=np.arange(len(years))
width=0.2
plt.bar(x,kohli,width,label='Kohli')
plt.bar(x-width,sachin,width,label='Sachin')
plt.bar(x+width,sehwag,width,label='Sehwag')
plt.title('Runs Comparison Over Years')
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.tight_layout()
plt.xticks(x,years)
plt.show()
```

```python
players = ['Sehwag','Kohli','Sachin']
debut = [2300,2100,2500]
plt.barh(players, debut,color=['blue','green','orange'])
plt.xlabel('Total Runs')
plt.title('Runs in Starting 5 Years')
plt.grid()
plt.show()
```

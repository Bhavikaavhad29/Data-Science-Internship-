```python

import matplotlib.pyplot as plt
import numpy as np

# Sample x values
x = np.linspace(0, 10, 100)

# Different y-values for variety
y_values = [np.sin(x + i * 0.5) for i in range(10)]

# Get 10 different matplotlib styles
styles = plt.style.available[:10]

# Different colors, markers, and linestyles
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k', '#ff7f0e', '#2ca02c', '#9467bd']
markers = ['o', 's', '^', 'D', '*', 'x', '+', 'v', '>', '<']
linestyles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--']

# Create a figure with 10 subplots
fig, axes = plt.subplots(5, 2, figsize=(12, 12))
axes = axes.flatten()

for i in range(10):
    plt.style.use(styles[i])
    axes[i].plot(x, y_values[i], 
                 color=colors[i], 
                 marker=markers[i], 
                 linestyle=linestyles[i], 
                 label=f'Style: {styles[i]}')
    axes[i].set_title(f'Style: {styles[i]}')
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

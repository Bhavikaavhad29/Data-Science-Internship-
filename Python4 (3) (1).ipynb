```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
plt.style.available
```

```python
Brand_A=[24,35,76,55,77,33,88]
Brand_B=[45,67,89,34,86,9,2]
days=[2015, 2016, 2017, 2018, 2019, 2020, 2021]
plt.style.use("seaborn-v0_8-bright")
plt.figure(figsize=(10,6))
plt.plot(days,Brand_A,label="Brand_A")
plt.plot(days,Brand_B,label="Brand_B")
plt.xlabel("Years")
plt.ylabel("Mobile Phones sold")
plt.title("Mobile Phones served in 7 days")
plt.legend()
plt.show()
```
